```python
#테스트용 100개 이미지
car_images = car_images[:100]
noncar_images = noncar_images[:100]
imgs = car_images + noncar_images
```

```python
#파라미터들 의 값
#svc 머신러닝 - 이미 선언됨
#X_scaler - 머신러닝 부분인데 물어보기
#HOG 알고리즘 파라미터
orient = 9
pix_per_cell = (4, 4)
cell_per_block = (2, 2)
#이미지 리사이즈 파라미터
spatial = (32, 32)
#히스토그램 파라미터
hist_bins = [32]
```

```python
"""
이미지 리사이즈
"""
def img_resize(img, spatial):
    src = cv2.imread(img)
    src_resize = cv2.resize(src, spatial) # fx: 가로 사이즈의 배수
    resize_feature = src_resize.flatten()

    return resize_feature
```

```python
"""
hog 알고리즘 
"""
def hog_img(img, orient, pix_per_cell, cell_per_block):
    hog_img = mpimg.imread(img)
    gray_img = cv2.cvtColor(hog_img, cv2.COLOR_BGR2GRAY)
    
    hog_feature, hog_image = hog(gray_img, orientations=orient, pixels_per_cell=pix_per_cell, 
                                     cells_per_block=cell_per_block, block_norm='L2-Hys', 
                                     transform_sqrt=False, visualize = True, 
                                     feature_vector=False)
    
    hog_feature = hog_feature.flatten()
        
    return hog_feature
```

```python
"""
히스토그램 분석
"""
def hist_img(img, hist_bins ):
    
    src = cv2.imread(img)
    hist1_b = cv2.calcHist(images = [src], 
                         channels = [0], 
                         mask = None, 
                         histSize = hist_bins, 
                         ranges = [0, 255])

    hist2_g = cv2.calcHist(images = [src], 
                         channels = [1], 
                         mask = None, 
                         histSize = hist_bins, 
                         ranges = [0, 255])

    hist3_r = cv2.calcHist(images = [src], 
                         channels = [2], 
                         mask = None, 
                         histSize = hist_bins, 
                         ranges = [0, 255])

    hist1_b_f = hist1_b.flatten()
    hist2_g_f = hist2_g.flatten()
    hist3_r_f = hist3_r.flatten()

    hist_feature = np.concatenate([hist1_b_f, hist2_g_f, hist3_r_f]) # 특성 합치는 함수

    return hist_feature
```

```python
"""
레이블 - 머신러닝 정답 
"""
label_list = []
for i in range(len(car_images)):
    answer=1
    label_list.append(answer)

    
for i in range(len(noncar_images)):
    notanswer=0
    label_list.append(notanswer)
```

```python
labeled = np.array(label_list)
```

```python
feature_list = []
for file in imgs:
    feature1 = img_resize(file, spatial )
    feature2 = hog_img(file, orient, pix_per_cell, cell_per_block,)
    feature3 = hist_img(file, hist_bins )
    feature = np.concatenate((feature1, feature2, feature3))
    feature_list.append(feature)
X = np.array(feature_list)
```
